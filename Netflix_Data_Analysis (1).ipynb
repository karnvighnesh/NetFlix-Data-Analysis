```python
# Ensure 'date_added' is in datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
recent_years_data = df[df['date_added'].dt.year >= (2025 - 10)]
recent_years_data['year_added'] = recent_years_data['date_added'].dt.year
content_by_year = recent_years_data.groupby(['year_added', 'type']).size().reset_index(name='count')
plt.figure(figsize=(12, 6))
sns.lineplot(data=content_by_year, x='year_added', y='count', hue='type', marker='o')
plt.title('Movies vs. TV Shows Added to Netflix Over the Last 10 Years', fontsize=16)
plt.xlabel('Year Added', fontsize=14)
plt.ylabel('Number of Titles Added', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Type')
plt.show()
```

```python
content_by_country = df.groupby('country').size().reset_index(name='count')
content_by_country = content_by_country.sort_values(by='count', ascending=False)
content_by_country = content_by_country.reset_index(drop=True)
print("Top 10 Countries by Content Count:")
print(content_by_country.head(10))
```

```python
total_content = content_by_country['count'].sum()
content_by_country['percentage'] = (content_by_country['count'] / total_content) * 100
top_10_countries_percentage = content_by_country.head(10).reset_index(drop=True)
```

```python
# Ensure 'date_added' is in datetime format
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Filter only TV shows and remove rows with NaT in 'date_added'
tv_shows = df[df['type'] == 'TV Show']
tv_shows = tv_shows.dropna(subset=['date_added'])

# Extract month and year
tv_shows['month_added'] = tv_shows['date_added'].dt.month
tv_shows['year_added'] = tv_shows['date_added'].dt.year

# Group by month and count
shows_by_month = tv_shows['month_added'].value_counts().sort_index()

print("Number of TV Shows Added by Month:")
print(shows_by_month)
```

```python
shows_by_month = tv_shows['month_added'].value_counts().sort_index()

# Display the count of TV shows added per month
print("Number of TV Shows Added by Month:")
print(shows_by_month)
```

```python
del range
plt.figure(figsize=(10, 6))
sns.barplot(x=shows_by_month.index, y=shows_by_month.values, palette='viridis')

# Add labels and formatting
plt.title('Number of TV Shows Added by Month', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of TV Shows', fontsize=14)
plt.xticks(
    ticks=range(1, 13),  # Months are 1 to 12
    labels=[
        'January', 'February', 'March', 'April', 'May', 'June',
        'July', 'August', 'September', 'October', 'November', 'December'
    ],
    rotation=45
)
plt.show()
```

```python
# Group by 'country' and 'type', and count the number of titles
country_type_count = df.groupby(['country', 'type']).size().reset_index(name='count')

# Separate the data for movies and TV shows
movies_by_country = country_type_count[country_type_count['type'] == 'Movie']
tv_shows_by_country = country_type_count[country_type_count['type'] == 'TV Show']

# Find the country with the highest number of movies
top_movie_country = movies_by_country.sort_values(by='count', ascending=False).head(1)

# Find the country with the highest number of TV shows
top_tv_show_country = tv_shows_by_country.sort_values(by='count', ascending=False).head(1)

# Display results
print("Country Producing the Highest Number of Movies:")
print(top_movie_country)

print("\nCountry Producing the Highest Number of TV Shows:")
print(top_tv_show_country)
```

```python
movies = df[df['type'] == 'Movie']
tv_shows = df[df['type'] == 'TV Show']

# Extract numeric values for Movies (duration in minutes)
movies['duration_minutes'] = movies['duration'].str.extract('(\d+)').astype(float)

# Extract numeric values for TV Shows (number of seasons)
tv_shows['num_seasons'] = tv_shows['duration'].str.extract('(\d+)').astype(float)

# Display samples
print("Movies Duration (Minutes):")
print(movies[['title', 'duration', 'duration_minutes']].head())

print("\nTV Shows Duration (Seasons):")
print(tv_shows[['title', 'duration', 'num_seasons']].head())
```

```python
top_countries=movies['country'].value_counts().head(10).index
filtered_df=movies[df['country'].isin(top_countries)]
plt.figure(figsize=(12, 6))
sns.boxplot(x='country',y='duration_minutes',data=filtered_df,palette='viridis')
plt.title('Movie Durations by Country',fontsize=16)
plt.xlabel('Country',fontsize=14)
plt.ylabel('Duration (Minutes)',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
continous_cols=['duration_minutes','release_year']
corr_matrix=movies[continous_cols].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correaltion Heatmap',fontsize=16)
plt.show()
```

```python
selected_columns=['duration_minutes','release_year','type']
sns.pairplot(movies[selected_columns],hue='type',palette='viridis',diag_kind='kde')
plt.show()
```

```python
df['year_added']=df['date_added'].dt.year
content_trend=df.groupby(['year_added','type']).size().reset_index(name='count')
plt.figure(figsize=(12,6))
sns.lineplot(data=content_trend,x='year_added',y='count',hue='type',marker='o')
plt.title('Trend of Movies vs Tv Shows Added Over the Years',fontsize=16)
plt.xlabel('Year Added',fontsize=14)
plt.ylabel('Number of Titles Added',fontsize=14)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.legend(title='Type')
plt.show()
```

```python
df_exploded=df.assign(listed_in=df['listed_in'].str.split(',')).explode('listed_in')
df_exploded['listed_in']=df_exploded['listed_in'].str.strip()
top_genres=df_exploded['listed_in'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_genres.index,y=top_genres.values,palette='viridis')
plt.title('Top 10 Genres',fontsize=16)
plt.xlabel('Number of Titles',fontsize=14)
plt.ylabel('Genre',fontsize=14)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df,x='rating',order=df['rating'].value_counts().index,palette='viridis')
plt.title('Distribution of Ratings',fontsize=16)
plt.xlabel('Rating',fontsize=14)
plt.ylabel('Number of Titles',fontsize=14)
plt.xticks(rotation=45)
plt.show()
```

```python
country_genre=df_exploded.groupby(['country','listed_in']).size().reset_index(name='count')
top_countries = df['country'].value_counts().head(5).index
filtered_data = country_genre[country_genre['country'].isin(top_countries)]
plt.figure(figsize=(14, 8))
genre_pivot = filtered_data.pivot(index='country', columns='listed_in', values='count').fillna(0)
sns.heatmap(genre_pivot, cmap='coolwarm', annot=False)
plt.title('Most Common Genres by Country', fontsize=16)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Country', fontsize=14)
plt.show()
```
